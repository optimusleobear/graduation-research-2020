# Introducing RealSR

RealSR stands for Real-World Super-Resolution via Kernel Estimation and Noise Injection (CVPR 2020 Workshops).

It's a project from Xiaozhong Ji, Yun Cao, Ying Tai, Chengjie Wang, Jilin Li, and Feiyue Huang, Tencent *Youtu Lab*.

Details can be found on its [GitHub Project Page](https://github.com/Tencent/Real-SR).

## Real-SR Models

Information of pre-trained models are as below.

**Models for challenge results**

- DF2K for corrupted images with processing noise.
- DPED for real images taken by cell phone camera.

**Extended models**

- DF2K-JPEG for compressed jpeg image.
